```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table between 'Type' and 'Genre'
contingency_table = pd.crosstab(df['BHK'], df['Rent'])

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Interpret the results (e.g., check if p < 0.05 for significance)
if p < 0.05:
    print("There is a significant association between Type and Genre.")
else:
    print("There is no significant association between Type and Genre.")
```

The chi-square (χ²) statistic measures the association between two categorical variables in a contingency table. It is used to determine if there is a significant association (dependency) between the two variables. The formula for the chi-square statistic is as follows:

Chi-Square (χ²) = Σ [(O - E)² / E]

Where:

Σ represents the sum over all cells in the contingency table.
O represents the observed frequency in a particular cell of the table.
E represents the expected frequency in the same cell if there were no association between the variables.
