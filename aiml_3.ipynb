```python
import matplotlib.pyplot as pyplot

pyplot.bar([0.25,2.25,3.25,5.25,7.25],[300,400,200,600,700],
label="area",color='b',width=0.5)
pyplot.bar([0.75,1.75,2.75,3.75,4.75],[50,30,20,50,60],
label="locality", color='g',width=.5)
pyplot.legend()
pyplot.xlabel('rent')
pyplot.ylabel('size')
pyplot.title('Details')

# Print the chart
pyplot.show()
```

```python
import matplotlib.pyplot as pyplot

slice = [19.6,23.4,35,21,1]
activities = ['kolkata','salt lake', 'bandhel', 'south dumdum', 'fool bagan']
cols = ['r','b','c','g', 'orange']
pyplot.pie(slice,
labels =activities,
colors = cols,
startangle = 90,
explode =(0,0.1,0,0,0),
autopct ='%1.1f%%')
pyplot.title('population')
pyplot.show
```
